create a data frame manually

```python
import pandas as pd

data={
    'name': ['alice','bob','charlie','david','eva'],
    'age': [25,30,35,40,22],
    'department': ['sales','engineerting','hr','engineering','sales'],
    'salary': [50000,80000,40000,90000,45000],
}
df=pd.DataFrame(data)
df
```

basic operations


```python
print("first 3 rows:")
df.head(3)
```

```python
print("column names")
df.columns
```

```python
print("data type:")
df.dtypes
```

```python
print("describe numerical columns:")
df.describe()
```

filtering

```python
print("people with salary > 50000:")
df[df['salary'] > 50000]
```

sorting

```python
print("sorted by age:")
df.sort_values(by='age')
```

grouping and aggeregation

```python
grouped = df.groupby('department')['salary'].mean()
print("average salary by department:")
grouped
```

adding new columns

```python
df['salary in lakhs'] = df['salary']/100000
df
```

line chart

```python
import matplotlib.pyplot as plt

years = [2019,2020,2021,2022]
revenue = [100,120,130,160]

plt.plot(years,revenue,marker='o', linestyle='-', color='blue')
plt.title("company revenue over years")
plt.xlabel("year")
plt.ylabel("revenue(in crore)")
plt.xticks(years)
plt.grid(True)
plt.show()



```

bar chart

```python
departments = df['department'].value_counts()
departments.plot(kind='bar',color='teal')
plt.title("employee count by department")
plt.xlabel("department")
plt.ylabel("count")
plt.show()
```

histogram

```python
ages = df['age']
plt.hist(ages, bins=5,color='aqua',edgecolor='black')
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("frequency")
plt.show()
```

pie chart

```python
df.groupby('department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("department proportion")
plt.ylabel("")
plt.show()
```

load sample dataset

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
print("\n info about the tips datasets:")
tips.info()

print("\n description of the tips dataset:")
tips.describe(include='all')
```

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("correlation matrix")
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.show()
```

```python
sns.countplot(x='day', data=tips, palette='Set2', hue='day',legend=False)
plt.title("count of records by day")
plt.show()
```

interpreting the boxplot

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("boxplot of total bill by day")
plt.show()
```
